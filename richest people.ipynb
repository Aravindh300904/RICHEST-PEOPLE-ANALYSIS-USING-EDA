```python
#1.SHOW THE AGE DISTRIBUTION AMONG THE DATA USING BAR PLOT
sns.histplot(data=df, x='age', bins=20, kde=True, color='skyblue')

# Customize the plot
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')

# Show the plot
plt.show()
```

```python
#2.SHOW THE NETWORTH VS FREQUENCY USING BAR PLOT
sns.histplot(data=df, x='networth', bins=20, kde=True, color='salmon')

# Customize the plot
plt.title('Net Worth Distribution')
plt.xlabel('Net Worth')
plt.ylabel('Count')

# Show the plot
plt.show()
```

```python
#3.SHOW INDUSTRY VS FREQUENCY USING BAR PLOT
plt.figure(figsize=(10, 6))
sns.set(style='whitegrid')  # Optional: Set a seaborn style

# Create the bar plot
sns.countplot(data=df, x='industry', palette='viridis')

# Customize the plot
plt.title('Industry Distribution')
plt.xlabel('Industry')
plt.ylabel('Count')

# Rotate x-axis labels if needed
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.show()
```

```python
#4. SHOW HOW DOES NET WORTH CHANGE WITH AGE AND INDUSTRY USING CAT PLOT
plt.figure(figsize=(10, 6))

# Create the cat plot
sns.catplot(data=df, x='age', y='networth', hue='industry', kind='bar', height=6, aspect=1.5, palette='Set2')

# Customize the plot
plt.title('Net Worth by Age and Industry')
plt.xlabel('Age')
plt.ylabel('Net Worth')

# Show the plot
plt.show()
```

```python
#5.SHOW THE TOP 10 RICHEST PEOPLE VS NET WORTH
names = [
    "Elon Musk", "Jeff Bezos", "Larry Ellison", "Mark Zuckerberg",
    "Bernard Arnault", "Warren Buffett", "Bill Gates", "Larry Page",
    "Amancio Ortega", "Sergey Brin"
]
net_worths = [
    251.5, 204.6, 203.7, 185.4, 172.4, 143.4, 138.5, 132.5, 128.0, 126.9
]

# Display the top 10 richest people
print("Top 10 Richest People:")
for rank, (person, net_worth) in enumerate(zip(names, net_worths), start=1):
    print(f"{rank}. {person}: ${net_worth:.1f} billion")
```

```python
#6.SHOW THE RICHEST PEOPLE FRON INDIA WITH THE NAMES IN ANY PLOT

# Data for the top richest people in India
names = [
    "Mukesh Ambani", "Gautam Adani", "Shiv Nadar", "Cyrus Poonawalla",
    "Lakshmi Mittal", "Savitri Jindal", "Dilip Shanghvi", "Radhakishan Damani",
    "Kumar Mangalam Birla", "Uday Kotak"
]
net_worths = [
    83.4, 47.2, 25.6, 22.6, 17.7, 17.5, 15.6, 15.3, 14.2, 12.9
]

# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(names, net_worths, color='skyblue')
plt.xlabel('Richest People in India')
plt.ylabel('Net Worth (in billions of dollars)')
plt.title('Top Richest People in India (2023)')
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#7.SHOW THE MINIMUM AGE BILLIONAIRE <=50 WITH NAME AND INDUSTRY .

# Sample data (names, ages, net worths, and industries)
names = ["Lukas Walton", "Alexandra Andresen", "Gustav Magnar Witze", "Evan Spiegel", "Ben Francis"]
ages = [37, 25, 28, 33, 31]
net_worths = [83.4, 47.2, 28.6, 22.0, 15.6]
industries = ["Retail", "Investment", "Fish farming", "Technology", "Retail"]

# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(names, net_worths, color='skyblue')
plt.xlabel('Billionaire Name')
plt.ylabel('Net Worth (in billions of dollars)')
plt.title('Youngest Billionaires and Their Net Worths')
plt.xticks(rotation=45, ha='right')

# Annotate each bar with the corresponding industry
for i, name in enumerate(names):
    plt.annotate(industries[i], xy=(name, net_worths[i]), xytext=(5, 5), textcoords="offset points", ha='center')

# Show the plot
plt.tight_layout()
plt.show()
```
